```python
missing_values = ["n/a", "na", "--", "-"]

#dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
#df = pd.read_csv(infile, parse_dates=['datetime'], date_parser=dateparse)

sales = pd.read_csv("\eksikveriler.csv",na_values = missing_values)
```

```python
cols = sales.columns
cols
```

```python
obj_cols=[]

for i in cols:
    
    col_type = sales[i].dtypes
    
    if col_type == 'object' or col_type == 'bool':
        obj_cols.append(i)

print('String veri türü kolonlar')        
print(obj_cols)
```

```python
num_cols=[]

for i in cols:
    
    col_type = sales[i].dtypes
    
    if col_type == 'int64' or col_type == 'float64':
        num_cols.append(i)
print('Sayı veri türü kolonlar') 
print(num_cols)
```

```python
# Detecting numbers 
cnt=0
for i in obj_cols:
    for row in sales[i]:
        try:
            int(row) or float(row)
            print('String veriler içinde sayı tespit edildi')
            #print(i, cnt)
            sales.loc[cnt, i]=np.nan
            print('Sayılar NaN olarak değiştirildi')
        except ValueError:
            pass
        cnt+=1
    cnt=0
```

```python
def isThereNull(liste):
    
    flag =False
    check_list = list(liste.isnull().sum())
    NA_col_names = []
    
    for i in range(len(check_list)):
        if check_list[i] != 0:
            col_name = cols[i]
            col_type = sales[col_name].dtypes
            NA_col_names.append(col_name)
            flag = True
    
            #print("Null değerli kolon indexi {}, kolon adı {}, kolon tipi {} " .format(i,col_name,col_type))
            
            
    if flag == False:
        print('Null yok')
    #print(NA_col_names)
    return NA_col_names
```

```python
if sales.isnull().values.any():
    NA_col_names=isThereNull(sales)
    print('Eksik veri bulunan kolonlar')
    print(NA_col_names)
```

```python
def fillNulls():
    NA_col_names = isThereNull(sales)
    
    for i in NA_col_names:
        col_type = sales[i].dtypes
        #print(col_type)
        if col_type == 'float64' or col_type == 'int64':
            print('Eksik sayılar ort değer ile dolduruldu.')
            mean = sales[i].mean()
            sales[i].fillna(mean, inplace=True)
           
        elif col_type =='object':
            print('Eksik str veriler No Values olarak değiştirildi.')
            sales[i].fillna('No Value', inplace=True)
            
```

```python
fillNulls()
```
